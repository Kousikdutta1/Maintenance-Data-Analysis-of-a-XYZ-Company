#**Project Name - Maintenance Data Analysis of a XYZ Company.**

```python
# Lt's first load different important library.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/RE Sustainability Maintenance Data Analysis/Breakdown status.csv'
```

```python
breakdown_vehicles = pd.read_csv(file_path)
```

```python
# Let's view the dataset

breakdown_vehicles
```

##**Let's plot a line plot to understand Percentage of Breakdown for the past 6 month.**

```python
plt.figure(figsize=(12,8))
sns.set(style="darkgrid")
sns.lineplot(x = 'Month', y = '% Breakdown', data = breakdown_vehicles);
plt.title("Percentage of Breakdown According to Month", fontsize = 13)
plt.xlabel("Month -->", fontsize = 13)
plt.ylabel("Percentage of Breakdown --> ", fontsize = 13)
plt.show();
```

**From above data it can be said that maximum breakdown of fleet was in April month, where breakdown of fleet was almost 8 %, which is higher than the authorized parameter(4-5%).**

#**Let's plot a line plot to understand Percentage of Availability of Fleet for the past 6 month.**

```python
plt.figure(figsize=(12,8))
sns.set(style="whitegrid")
sns.lineplot(x = 'Month', y = '% Availability', data = breakdown_vehicles);
plt.title("Percentage of Availability According to Month", fontsize = 13)
plt.xlabel("Month -->", fontsize = 13)
plt.ylabel("Percentage of Availability of Fleet --> ", fontsize = 13)
plt.show();
```

**From above data it can be said that maximum Availability of fleet was in February month.**

#**Let's plot a line plot to understand Repair and Maintenance Cost for the past 6 month.**


```python
plt.figure(figsize=(12,8))
sns.set(style="darkgrid")
sns.lineplot( x = 'Month',y = 'Repair and Maintenance Cost', data = breakdown_vehicles, color = 'red');
plt.title("Repair and Maintenance Cost According to Month", fontsize = 13)
plt.xlabel("Month -->", fontsize = 13)
plt.ylabel("Repair and Maintenance Cost --> ", fontsize = 13)
plt.show();
```

**From above data it can be said that maximum Repair and Maintenance Cost of fleet was in May month.**
